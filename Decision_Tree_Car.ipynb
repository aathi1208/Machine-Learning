```python
data['buying'],_=pd.factorize(data['buying'])#factorizing all the values into numerical=label encoding
data['maint'],_=pd.factorize(data['maint'])#'_' is important
data['doors'],_=pd.factorize(data['doors'])
data['persons'],_=pd.factorize(data['persons'])
data['lug_boot'],_=pd.factorize(data['lug_boot'])
data['safety'],_=pd.factorize(data['safety'])
data.head()
```

```python
 x=data.iloc[:,0:-1]#without class
 y=data.iloc[:,-1]#class
 x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#30 per test and 70 per train and random_state=0 to take continous value
#always take random_state as 42 to take random values
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)#accuracy for train data
print('Test Accuracy:{:.2f}'.format(accuracy))
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_test,y_test)
```

```python
y_train_pred=dtree.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_train_pred)#accuracy for test data
print('Test Accuracy:{:.2f}'.format(accuracy))
```
