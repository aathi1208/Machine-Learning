```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files #Take the Dataset From Kaggle Dataset name Iris
uploaded = files.upload()
```

```python
import io
df=pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.set_index("Id",inplace=True)
df
```

```python
df.dtypes
```

```python
df['Species'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Species']=lb.fit_transform(df['Species'])
```

```python
X_train=df['Species']
```

```python
df.drop(labels="Species",axis=1,inplace=True)
```

```python
X_train
```

```python
X_test=df
```

```python
from sklearn.model_selection import train_test_split
X_train_sub,X_validation_sub,y_train_sub,y_validation_sub=train_test_split(X_test,X_train,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
learning_rates=[0.05,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
  gb=AdaBoostClassifier(n_estimators=20,random_state=0)
  gb.fit(X_train_sub,y_train_sub)
  print("Learning rate: ",learning_rate)
  print("Accuracy score (training): {0:.3f}".format(gb.score(X_train_sub,y_train_sub)))
  print("Accuracy score (Validation): {0:.3f}".format(gb.score(X_validation_sub,y_validation_sub)))
  print("")
```
